```python
from sklearn.compose import ColumnTransformer

def preprocess(df):
    """
    Fills missing values and turns categorical columns to numbers
    df: DataFrame to preprocess
    """
    # Create the imputer objects
    num_imputer = SimpleImputer(strategy="mean")
    cat_imputer = SimpleImputer(strategy="most_frequent")
    
    # Create the onehot encoder object
    onehot = OneHotEncoder()
    
    # Identify the numkkmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmmerical and categorical columns
    num_cols = [col for col in df.columns if np.issubdtype(df[col].dtype, np.number)]
    cat_cols = [col for col in df.columns if not np.issubdtype(df[col].dtype, np.number)]
    other_cols = [col for col in df.columns if col not in num_cols and col not in cat_cols]
    
    # Create column transformer
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', num_imputer, num_cols),
            ('cat', onehot, cat_cols)
        ], 
        remainder='passthrough'
    )
    
    # Apply the transformers
    df = pd.DataFrame(preprocessor.fit_transform(df), columns=df.columns)
    return df
```
